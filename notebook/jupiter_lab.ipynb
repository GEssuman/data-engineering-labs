### Making an Api Request

```python
import requests
import pandas as pd
import numpy as np
import matplotlib as plt
import os

```

```python
BASE_URL = "https://api.themoviedb.org/3/movie/"
API_ACCESS_TOKEN = os.environ.get('API_ACCESS_TOKEN')
HEADERS = {
    "accept": "application/json",
    "Authorization": f"Bearer {API_ACCESS_TOKEN}"
}
movie_ids = [0, 299534, 19995, 140607, 299536, 597, 135397, 420818, 24428, 168259, 99861, 284054, 12445,181808, 330457, 351286, 109445, 321612, 260513]
```

```python
movies = []
for id in movie_ids:
    url = f"{BASE_URL}{id}"
    response = requests.get(url, headers=HEADERS)
    print(url)
    if response.status_code == 200:
        data = response.json()
        print("Success")
        movie = {
            "id": data["id"],
            "title": data["title"],
            "release_date": data["release_date"],
            "popularity": data["popularity"],
            "vote_average": data["vote_average"],
            "vote_count": data["vote_count"],
            "genres": "|".join([genre["name"] for genre in data["genres"]]),
            "spoken_languages": "|".join([lang["english_name"] for lang in data["spoken_languages"]]),
            "production_countries": "|".join([country["name"] for country in data["production_countries"]])
        }
    else:
        print(f"Error occured trying to request movie with id :{id}\nError Status Code:{response.status_code}")

    movies.append(movie)



pd.DataFrame(movies)
```
